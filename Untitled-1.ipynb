```python
import tkinter as tk
from tkinter import messagebox
import pyodbc

# Database configuration
server = 'localhost'
database = 'Shahre ketab'
username = 'DESKTOP-1BU91CO\Alireza'
password = ''
driver = '{ODBC Driver 17 for SQL Server}'

# Establish database connection
def connect_to_db():
    try:
        conn = pyodbc.connect(f'DRIVER={driver};SERVER={server};DATABASE={database};UID={username};PWD={password}')
        return conn
    except Exception as e:
        messagebox.showerror("Database Connection Error", str(e))
        return None

# Search books by name
def search_books():
    name = entry_search.get()
    conn = connect_to_db()
    if conn:
        cursor = conn.cursor()
        query = "SELECT * FROM dbo.ketab WHERE onvan LIKE ?"
        cursor.execute(query, f'%{name}%')
        books = cursor.fetchall()
        conn.close()
        
        listbox_results.delete(0, tk.END)
        for book in books:
            listbox_results.insert(tk.END, f"Title: {book.onvan}, Writer1: {book.writer1}, Writer2: {book.writer2}, ISBN: {book.shabak}")

# Insert a new book
def insert_book():
    onvan = entry_onvan.get()
    writer1 = entry_writer1.get()
    writer2 = entry_writer2.get()
    shabak = entry_shabak.get()
    nasher = entry_nasher.get()
    gheymat = entry_gheymat.get()
    mozue = entry_mozue.get()
    nobat_chap = entry_nobat_chap.get()
    
    if not onvan or not writer1 or not shabak:
        messagebox.showwarning("Input Error", "Title, Writer1, and ISBN are required")
        return
    
    conn = connect_to_db()
    if conn:
        cursor = conn.cursor()
        query = "INSERT INTO dbo.ketab (onvan, writer1, writer2, shabak, nasher, gheymat, mozue, nobat_chap) VALUES (?, ?, ?, ?, ?, ?, ?, ?)"
        try:
            cursor.execute(query, onvan, writer1, writer2, shabak, nasher, gheymat, mozue, nobat_chap)
            conn.commit()
            messagebox.showinfo("Success", "Book inserted successfully")
            entry_onvan.delete(0, tk.END)
            entry_writer1.delete(0, tk.END)
            entry_writer2.delete(0, tk.END)
            entry_shabak.delete(0, tk.END)
            entry_nasher.delete(0, tk.END)
            entry_gheymat.delete(0, tk.END)
            entry_mozue.delete(0, tk.END)
            entry_nobat_chap.delete(0, tk.END)
        except Exception as e:
            messagebox.showerror("Database Error", str(e))
        finally:
            conn.close()

# Tkinter GUI setup
root = tk.Tk()
root.title("Book Management")

# Search books section
tk.Label(root, text="Search Books").grid(row=0, column=0, padx=10, pady=10)
entry_search = tk.Entry(root)
entry_search.grid(row=0, column=1, padx=10, pady=10)
btn_search = tk.Button(root, text="Search", command=search_books)
btn_search.grid(row=0, column=2, padx=10, pady=10)

listbox_results = tk.Listbox(root, width=80, height=10)
listbox_results.grid(row=1, column=0, columnspan=3, padx=10, pady=10)

# Insert book section
tk.Label(root, text="Insert New Book").grid(row=2, column=0, padx=10, pady=10)

tk.Label(root, text="Title").grid(row=3, column=0, padx=10, pady=5)
entry_onvan = tk.Entry(root)
entry_onvan.grid(row=3, column=1, padx=10, pady=5)

tk.Label(root, text="Writer1").grid(row=4, column=0, padx=10, pady=5)
entry_writer1 = tk.Entry(root)
entry_writer1.grid(row=4, column=1, padx=10, pady=5)

tk.Label(root, text="Writer2").grid(row=5, column=0, padx=10, pady=5)
entry_writer2 = tk.Entry(root)
entry_writer2.grid(row=5, column=1, padx=10, pady=5)

tk.Label(root, text="ISBN").grid(row=6, column=0, padx=10, pady=5)
entry_shabak = tk.Entry(root)
entry_shabak.grid(row=6, column=1, padx=10, pady=5)

tk.Label(root, text="Publisher").grid(row=7, column=0, padx=10, pady=5)
entry_nasher = tk.Entry(root)
entry_nasher.grid(row=7, column=1, padx=10, pady=5)

tk.Label(root, text="Price").grid(row=8, column=0, padx=10, pady=5)
entry_gheymat = tk.Entry(root)
entry_gheymat.grid(row=8, column=1, padx=10, pady=5)

tk.Label(root, text="Subject").grid(row=9, column=0, padx=10, pady=5)
entry_mozue = tk.Entry(root)
entry_mozue.grid(row=9, column=1, padx=10, pady=5)

tk.Label(root, text="Edition").grid(row=10, column=0, padx=10, pady=5)
entry_nobat_chap = tk.Entry(root)
entry_nobat_chap.grid(row=10, column=1, padx=10, pady=5)

btn_insert = tk.Button(root, text="Insert", command=insert_book)
btn_insert.grid(row=11, column=0, columnspan=2, padx=10, pady=10)

root.mainloop()
```
